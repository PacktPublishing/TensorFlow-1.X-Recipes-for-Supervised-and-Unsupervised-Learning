# Using Keras for building a DNN

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import os

from pandas import read_csv
from sklearn.model_selection import train_test_split
%matplotlib inline
```

## 1. Load and prepare the data

```python
current_dir = os.getcwd()
dataset_path = os.path.join(os.getcwd(), os.pardir,'data','diamond_prices.csv')
diamonds = read_csv(dataset_path)
TARGET = 'price'
X_data = diamonds.iloc[:,1:].values
y_data = diamonds[TARGET].values

## Separating into training and testing
X_train, X_val, y_train, y_val = train_test_split(X_data, y_data, train_size=0.9, random_state=60)
```

## 2. Instantiate a Sequential model

```python
from keras.models import Sequential
reg_model = Sequential()
```

## 3. Build the DNN model using Dense layers

```python
from keras.layers import Dense
```

This is what we did in TensorFlow core:
~~~~
n_hidden1 = 128
n_hidden2 = 64
n_hidden3 = 32
n_outputs = 1

def DNN(inputs):
    hidden1 = fully_connected(inputs, n_hidden1)
    hidden2 = fully_connected(hidden1, n_hidden2)
    hidden3 = fully_connected(hidden2, n_hidden3)
    y_pred = fully_connected(hidden3, n_outputs, activation_fn=None)
    return tf.squeeze(y_pred)
~~~~

```python
n_inputs = X_train.shape[1]
n_hidden1 = 128
n_hidden2 = 64
n_hidden3 = 32
n_outputs = 1

# first hidden layer, you only need to set the input_dim for the first layer
reg_model.add(Dense(units=n_hidden1, activation='relu', input_dim=n_inputs))
# second hidden layer
reg_model.add(Dense(units=n_hidden2, activation='relu'))
# third hidden layer
reg_model.add(Dense(units=n_hidden3, activation='relu'))
# output layer # for activation: If you don't specify anything, no activation is applied
reg_model.add(Dense(units=n_outputs))
```

## 4. Compile the model

```python
reg_model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['mean_squared_error'])
```

## 5. Train the model

```python
N_EPOCHS = 400
BATCH_SIZE = 128
reg_model.fit(X_train, y_train, epochs=N_EPOCHS, batch_size=BATCH_SIZE)
```

## 6. Visualize/analyze the results of the model

```python
## Getting the predictions from the model
predictions = reg_model.predict(X_val).flatten()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x=predictions, y=y_val, s=0.5)
ax.set_xlabel('Predicted prices')
ax.set_ylabel('Observed prices')
ax.set_title("Predictions vs. Observed Values in the validation set");
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.hist(abs(y_val-predictions), bins=100, edgecolor='black')
ax.set_xlim(0, 4e3);
```
