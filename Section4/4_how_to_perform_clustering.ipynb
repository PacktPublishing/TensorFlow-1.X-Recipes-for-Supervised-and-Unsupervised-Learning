# How to perform clustering

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
from tensorflow.contrib.learn import KMeansClustering

%matplotlib inline
```

## 1. Load and prepare the data

```python
mnist = input_data.read_data_sets("MNIST_data/")
```

## 2. Normalize the data

```python
#from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#scaler.fit(mnist.train.images)
```

## 3. Define the number of clusters and training strategy

```python
N_CLUSTERS = 10
N_IMAGES = mnist.train.images.shape[0]
EPOCHS = 500
BATCH_SIZE = 500
STEPS = int(EPOCHS*N_IMAGES/BATCH_SIZE)
```

## 4. Instantiate the model (tf.contrib.learn.KMeansClustering)

```python
k_means = KMeansClustering(num_clusters=N_CLUSTERS,
                           kmeans_plus_plus_num_retries=-1,
                           model_dir = './kmeans')
```

## 5. Define the training input function

```python
def train_input_fn():
    X, _  = mnist.train.next_batch(BATCH_SIZE)
  #  X = scaler.transform(X)
    X = tf.constant(X, tf.float32, X.shape)
    return {'x': X}, None
```

## 6. Use the .fit method to train the model

```python
k_means.fit(input_fn=train_input_fn, steps=STEPS)
```

## Let's take a look at the clusters' centers

```python
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(12,5))
for i in range(N_CLUSTERS):
    ax.flatten()[i].imshow(k_means.clusters()[i].reshape((28,28)), cmap='gray')
```
